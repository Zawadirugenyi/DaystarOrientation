```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data =pd.read_csv("/content/students_adaptability_level_online_education.csv")
data.head()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt

# Assuming 'data' is a pandas DataFrame with columns 'Age' and 'Education_Level'
plt.plot(data['Age'], data['Education Level'], marker='o', linestyle='-', color='b')
plt.xlabel('Age')
plt.ylabel('Education Level')
plt.title('Line Chart of Age vs Education Level')
plt.grid(True)  # Add grid lines for better readability
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'data' is a pandas DataFrame with columns 'Age' and 'Education_Level'
sorted_data = data.sort_values(by='Age')  # Sort data by 'Age' for the line chart

plt.plot(sorted_data['Age'], sorted_data['Education Level'], marker='o', linestyle='-', color='b')
plt.xlabel('Age')
plt.ylabel('Education Level')
plt.title('Line Chart of Age vs Education Level')
plt.grid(True)  # Add grid lines for better readability
plt.show()

```

```python
import matplotlib.pyplot as plt

# Assuming 'data' is a pandas DataFrame with a column 'Education_Level'
education_counts = data['Education Level'].value_counts()

# Plotting a pie chart
plt.pie(education_counts, labels=education_counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle
plt.title('Distribution of Education Levels')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'data' is a pandas DataFrame with a column 'Education_Level'
plt.hist(data['Education Level'], bins=10, edgecolor='black')
plt.xlabel('Education Level')
plt.ylabel('Frequency')
plt.title('Histogram of Education Levels')
plt.show()
```
