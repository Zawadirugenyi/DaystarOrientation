```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data =pd.read_csv("/content/data_cleaned_2021.csv")
data.head()
```

```python
data.describe()
```

```python
plt.scatter(data.index,data.Rating)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'data' is a pandas DataFrame with a column 'Rating'
plt.hist(data['Rating'], bins=10, edgecolor='black')  # You can adjust the number of bins as needed
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Histogram of Ratings')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'data' is a pandas DataFrame with a column 'Rating'
ratings_counts = data['Rating'].value_counts()

# Plotting a pie chart
plt.pie(ratings_counts, labels=ratings_counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.title('Distribution of Ratings')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'data' is a pandas DataFrame with a column 'Rating' and an index suitable for time-based data
# Replace 'Rating' and 'your_time_column' with the actual column names from your DataFrame

plt.plot(data.index, data['Rating'], marker='o', linestyle='-', color='b')
plt.xlabel('Time or X-axis Label')
plt.ylabel('Rating')
plt.title('Line Chart of Ratings Over Time')
plt.grid(True)  # Add grid lines for better readability
plt.show()
```
